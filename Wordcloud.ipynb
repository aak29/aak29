```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
import sys, os
os.chdir(sys.path[0])

```

```python
# Read text
text = open('Tesla.txt', mode='r', encoding='utf-8').read()
stopwords = STOPWORDS
```

```python
#create graphics
#light version
wc = WordCloud(
        background_color='white',
        stopwords=stopwords,
        height = 600,
        width=400
)

wc.generate(text)

#dark version
wbc = WordCloud(
        background_color='black',
        stopwords=stopwords,
        height = 400,
        width=600
)
wbc.generate(text)
```

```python
# store to file
wc.to_file('TeslaWordCloud.png')
wbc.to_file('TeslaWordCloudDark.png')

#printout to the screen
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
