```python

#import the packages
import datetime as dt
import matplotlib.pyplot as plt
from matplotlib import style
import pandas as pd
import pandas_datareader.data as web
```

```python
#Pick a style:
style.use('ggplot')
```

```python
start = dt.datetime(2022,1,1)
end = dt.datetime(2022,12,31)
```

```python
#Create a dataframe
df_tsla = web.DataReader('TSLA', 'stooq', start, end)
df_ford= web.DataReader('F', 'stooq', start, end)
df_toyota= web.DataReader('TM', 'stooq', start, end)
```

```python
#flip dates

df_tsla2=df_tsla
df_tsla2 = df_tsla2.iloc[::-1]

df_ford2=df_ford
df_ford2=df_ford2.iloc[::-1]

df_toyota2=df_toyota
df_toyota2=df_toyota2.iloc[::-1]
```

```python
#Reduce decimal places for easier viewing
pd.set_option("display.precision", 2)
```

```python
df_tsla = pd.read_csv('tsla.csv', parse_dates=True, index_col=0)
df_ford = pd.read_csv('ford.csv', parse_dates=True, index_col=0)
df_toyota = pd.read_csv('toyota.csv', parse_dates=True, index_col=0)
```

```python
df_tsla['100ma'] = df_tsla['Close'].rolling(window=100, min_periods=0).mean()
df_tsla
```

```python
df_ford['100ma'] = df_ford['Close'].rolling(window=100, min_periods=0).mean()
df_ford
```

```python
df_toyota['100ma'] = df_toyota['Close'].rolling(window=100, min_periods=0).mean()
df_toyota
```

```python
#The size for our chart:
plt.figure(figsize = (12,6))
#Plotting Adj Close and SMA lines:
ax1 = plt.subplot2grid((8,1),(0,0),rowspan=5,colspan=1)
ax2 = plt.subplot2grid((8,1),(6,0),rowspan=3,colspan=1, sharex=ax1)

ax1.plot(df_tsla.index,df_tsla['Close'], 'r',label='Close')
ax1.plot(df_tsla.index,df_tsla['100ma'], 'g', label= "100 Moving Average")
plt.xlabel('Date')
#ax1.set_xlabel('Date')

ax1.set_ylabel('Closing price ($)')
ax1.set_title('Close and 100 day Moving Average for TSLA')
ax1.legend()

ax2.bar(df_tsla.index,df_tsla['Volume'], label = "Volume", color='b')
ax2.legend()

plt.show()
```

```python
#The size for our chart:
plt.figure(figsize = (12,6))
#Plotting Adj Close and SMA lines:
ax1 = plt.subplot2grid((8,1),(0,0),rowspan=5,colspan=1)
ax2 = plt.subplot2grid((8,1),(6,0),rowspan=3,colspan=1, sharex=ax1)

ax1.plot(df_ford.index,df_ford['Close'], 'r',label='Adjusted Close')
ax1.plot(df_ford.index,df_ford['100ma'], 'g', label= "100 Moving Average")
plt.xlabel('Date')
#ax1.set_xlabel('Date')
ax1.set_ylabel('Closing price ($)')
ax1.set_title('Close and 100 day Moving Average for Ford Motors')
ax1.legend()

ax2.bar(df_ford.index,df_ford['Volume'], label = "Volume", color='b')
ax2.legend()

plt.show()
```

```python
#The size for our chart:
plt.figure(figsize = (12,6))
#Plotting Adj Close and SMA lines:
ax1 = plt.subplot2grid((8,1),(0,0),rowspan=5,colspan=1)
ax2 = plt.subplot2grid((8,1),(6,0),rowspan=3,colspan=1, sharex=ax1)

ax1.plot(df_toyota.index,df_toyota['Close'], 'r',label='Adjusted Close')
ax1.plot(df_toyota.index,df_toyota['100ma'], 'g', label= "100 Moving Average")
plt.xlabel('Date')
#ax1.set_xlabel('Date')
ax1.set_ylabel('Closing price ($)')
ax1.set_title('Close and 100 day Moving Average for Toyota Motors')
ax1.legend()

ax2.bar(df_toyota.index,df_toyota['Volume'], label = "Volume", color='b')
ax2.legend()

plt.show()
```
