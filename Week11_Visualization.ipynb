```python
dataFile = 'tsla.csv'
#This creates a dataframe from the CSV file:
df = pd.read_csv(dataFile, index_col = 'Date')
df
```

```python
df.plot()
```

```python
close = df['Adj Close']
```

```python
close.index = pd.to_datetime(close.index)
close
```

```python
close.plot()
```

```python
df['100ma'] = close.rolling(window=100).mean()
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df['100ma'] = close.rolling(window=100, min_periods = 0 ).mean()
df
```

```python
sma= close.rolling(window=100, min_periods =0).mean()
sma
```

```python
sma.plot()
plt.title('100 Day Simple moving average');
```

```python
close.plot()
sma.plot()
```

```python
#The size for our chart:
plt.figure(figsize = (12,6))
#Plotting Adj Close and SMA lines:
plt.plot(close, label='TESLA Adj Close', linewidth = 2)
plt.plot(sma, label='100 day rolling SMA', linewidth = 1.5)

#Adding title and labels on the axes, making legend visible:
plt.xlabel('Date')
plt.ylabel('Adjusted closing price ($)')
plt.title('Price with a single Simple Moving Average')
plt.legend();

```

```python
ax1= plt.subplot2grid((8,1), (0,0), rowspan = 5, colspan =1) 
ax2= plt.subplot2grid((8,1), (6,0), rowspan = 1, colspan =1) 
```

```python
#The size for our chart:
plt.figure(figsize = (20,12))
#Plotting Adj Close and SMA lines:

ax1 = plt.subplot2grid((8,1),(0,0),rowspan=5,colspan=1)
ax2 = plt.subplot2grid((8,1),(6,0),rowspan=5,colspan=1, sharex=ax1)

ax1.plot(df.index,close, 'r', label='Adjusted Close')
ax1.plot(df.index,df['100ma'], 'g', label= "100 Moving Average")
ax1.set_title('Adjusted Close and 100 day Moving Average for TSLA')
ax1.set_ylabel('Adjusted closing price ($)')
ax1.legend()

ax2.bar(df.index,df['Volume'], label = "Volume", color='b')
ax2.legend()


plt.show()
```
